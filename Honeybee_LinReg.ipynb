# HoneyBee Linear Regression
#### In this analysis we will take the following CSV import from Kaggle to understand the decline in honeybee population and how this trend will predict future outcome of the honeybee population

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
from sklearn import linear_model

honeydf = pd.read_csv('honeyproduction.csv')
honeydf.head()
```

```python
prod_per_year = honeydf.groupby('year').mean('totalprod').reset_index()
prod_per_year.head()
```

```python
X = prod_per_year['year']
X = X.values.reshape(-1,1)
y = prod_per_year['totalprod']
print(X)
```

```python
plt.scatter(X,y)
plt.show()
plt.clf()
```

From this graph of minimal data points we can see a slight linear relationship between honey production and year. You can see that production declines from 1998 to 2012.

```python
regr = linear_model.LinearRegression()
regr.fit(X,y)
print(regr.coef_)
print(regr.intercept_)
y_predicted = regr.predict(X)
```

```python
plt.scatter(X,y)
plt.plot(X,y_predicted, color = 'orange')
plt.xlabel('Year')
plt.ylabel('Honey Production in lbs')
plt.title('Decline in Honey Production')
plt.show()
plt.clf()
```

Based on the Linear Regression above we will make future predictions of production levels in the year 2050.

```python
X_future = np.array(range(2013,2051))
X_future = X_future.reshape(-1,1)
futurepredict = regr.predict(X_future)
```

```python
plt.plot(X_future,futurepredict)
plt.xlabel('Year')
plt.ylabel('Honey Production in lbs')
plt.title('Predicted Decline in Honey Production')
plt.show()
plt.clf()
```
